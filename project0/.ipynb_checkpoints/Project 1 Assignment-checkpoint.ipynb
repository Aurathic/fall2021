```python
df = pandas.read_html(table_html)[0]
df.columns = ['rank', 'x_classification', 'date', 'region', 'start_time', 'maximum_time', 'end_time', 'movie']
df
```

## Step 2: Tidy the top 50 solar flare data

```python
df = df.drop('movie',axis=1)
# turn dates and times into datetimes
df['start_datetime'] = pandas.to_datetime(df['date']+ ' ' + df['start_time']) 
df['maximum_datetime'] = pandas.to_datetime(df['date']+ ' ' + df['maximum_time']) 
df['end_datetime'] = pandas.to_datetime(df['date']+ ' ' + df['end_time']) 
df = df.drop(['date', 'start_time', 'maximum_time', 'end_time'],axis=1)
# there are no missing values, so no extra filtering is needed
df
```

## Step 3: Scrape the NASA data

```python
# get NASA data
r2 = requests.get('https://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html')
soup2 = bs4.BeautifulSoup(r2.text, 'html.parser')
#print(soup2.prettify())
```

```python
# remove all links
for a in soup2.find_all('a'):
    a.replace_with_children()
# split string into array of rows, then turn into data table
lines2 = soup2.find('pre').text.split('\n')[12:-2]
# remove the final columns, which contain textual notes and 
# therefore don't fit in the table -- I also removed the `plot`
# column because it's useless without the link
lines2 = [line.split()[:14] for line in lines2]
df2 = pandas.DataFrame(lines2, columns = ['start_date', 'start_time', 'end_date', 'end_time', 'start_frequency', 'end_frequency', 'flare_location', 'flare_region', 'flare_classification', 'cme_date', 'cme_time', 'cme_angle', 'cme_width', 'cme_speed'])
df2
```

## Step 4: Tidy the NASA the table (sic)

```python
# replace missing data entries with Numpy's NaN
df2.replace('????',np.nan, inplace=True)
df2.replace(re.compile(r".*-.*"),np.nan, inplace=True)
# add column for whether row corresponds to Halo flare
# then remove Halo entries from cme_angle
df2['is_halo'] = df2.apply(lambda x:\
                            np.nan if pandas.isna(x['cme_angle']) else x['cme_angle'] == 'Halo', axis=1)
df2.replace('Halo', np.nan, inplace=True)
# add column for whether cme width is lower bound
df2['width_lower_bound'] = df2['cme_width'].map(
    lambda x: np.nan if pandas.isna(x) else x[0] == '>')
#    lambda x: pandas.notna(x) and x[0] == '>')
# Remove non-numeric characters (i.e the `>` in front of some `cme_width`s)
# Note: there's one value in `df2['cme_width']` which has an `h` on the end --
# I assume that's a typo, so I removed it from the end
# Note 2: I chose not to convert these into integers -- they're still strings
df2['cme_width'] = df2['cme_width'].map(lambda x: \
    re.sub("[^0-9]", "", x) if pandas.notna(x) else x)
df2[:]
```

```python
def make_datetime(date, time, year=None):
    """
    Takes in a date, time and an optional year (as a string), and turns that into
    a datetime object based on the rules. 
    """
    # return NaN if no date/time given
    if pandas.isna(date) or pandas.isna(time):
        return np.nan
    midnight = False
    ret_val = None
    # if time is 24:00, treat as 00:00 of next day
    if time == '24:00':
        time = '00:00' 
        midnight = True
    if year is None:
        ret_val = pandas.to_datetime(date + ' ' + time) 
    else:
        ret_val = pandas.to_datetime(year + '/' + date + ' ' + time)
    # accounting for 24:00 
    if midnight:
        ret_val += datetime.timedelta(days=1)
    return ret_val

# use function from above to get actual values
df2['start_datetime'] = df2.apply(lambda x: \
        make_datetime(date=x['start_date'], time=x['start_time']), axis=1)
# store year from start_date for use in end_datetime and cme_datetime
# this assumes that the end and cme date time take place in the same year as the start, which isn't always true 
# (e.g. if it starts at 1999/12/31 and ends 2000/01/01)
# that doesn't appear to occur in this data set though, so I decided to ignore it
df2['year'] = df2['start_datetime'].map(lambda x: str(x.year))
df2['end_datetime'] = df2.apply(lambda x: \
        make_datetime(date=x['end_date'], time=x['end_time'], year=x['year']), axis=1)
df2['cme_datetime'] = df2.apply(lambda x: \
        make_datetime(date=x['cme_date'], time=x['cme_time'], year=x['year']), axis=1)
# get rid of now-useless rows
df2 = df2.drop(['start_date', 'start_time', 'end_date', 'end_time', 'cme_date', 'cme_time', 'year'],axis=1)
df2
```

## Question 1: Replication

```python
# remove lone "FILA" -- I'm changing it to NaN because I don't know the intended classification
df2.at[66, 'flare_classification'] = np.nan
# extract flare class and magnitude (the letter, and the number afterwards)
df2['class'] = df2['flare_classification'].map(lambda x: \
                                              np.nan if pandas.isna(x) else x[0])
df2['magnitude'] = df2['flare_classification'].map(lambda x: \
                                              np.nan if pandas.isna(x) else float(x[1:]))
df2.sort_values(['class', 'magnitude'], ascending=[False, False])[:50]
# I am keeping the 'magnitude' and 'class' columns because I want to do data analysis on them
```

The top flares in this table are also included in in SpaceWeatherLive.com's Top 50 list, but a few flares shown in the Top 5 are not included in the table above - for example, an X17+ in 2005/09/07 and a X5.4 in 2005/09/08.

## Question 2: Integration

For a given flare in the Top 50 dataset, the flare in the NASA dataset with the closest start time is considered 'matching'. 
(This will cause every flare in the Top 50 to find a 'match' in the NASA dataset, even if the actual matching flare is not in the set. However, based on the project description, this seems to be the intended behavior.)

```python
#pandas.set_option('display.max_rows', None)

# get indices of values in df2['start_datetime'] which are closest to each value in df['start_time'] 
# adapted from https://stackoverflow.com/questions/30493614/pandas-merge-dataframes-based-on-closest-match
df['nasa_assoc_i'] = abs(np.atleast_2d(df['start_datetime']).T - df2['start_datetime'].values).argmin(axis=1)
# for each point in NASA dataset, find the index of the associated entry in the top 50 dataset.
# since these are supposedly the top 50 solar flares, the way I select between ties is take the flare in the WSL
# dataset that has the greatest ranking, because this situation rarely comes up in the actual data set 
# (only 2 times -- index 27 and 32), and this works well enough. 
# (assumes rankings should start at 1, rather than 0, and assumes that the flares are currently in 
# ranked order in the top 50 dataset. It also assumes that all of the flares in the ranked dataset are in 
# the NASA dataset, which does not appear to be true in reality, but I cannot think of a simple solution 
# to that issue.)

#print(list(df2.index.map(lambda x:\
#             len(np.where(df['nasa_assoc_i'] == x)[0]))))
#print(np.where(df['nasa_assoc_i'] == 32)[0])

df2['top_50_assoc_i'] = df2.index.map(lambda x:\
    np.where(df['nasa_assoc_i'] == x)[0][0]+1 \
    if len(np.where(df['nasa_assoc_i'] == x)[0]) > 0 \
    else np.nan)


df2
```

## Question 3: Analysis

```python
# data from the website linked in the README, for use of 
# calculating peak flux range
class_min_peak = {"A": -8, "B": -7, "C": -6, "M": -5, "X": -4}


def class_to_peak_flux_range(classification, magnitude):
    """
    Based on class and number, return the the peak flux
    range between 1 and 8 angstroms (watts per square meter).
    The way this works is based off of https://spaceweather.com/glossary/flareclasses.html
    (linked in the README).
    """
    return (np.nan
    if pandas.isna(classification) or pandas.isna(magnitude)
    else 10**class_min_peak[classification]) * magnitude

# add new column for peak flux range
df2['peak_flux_range'] = df2.apply(lambda x: \
    class_to_peak_flux_range(x['class'], x['magnitude']), axis=1)

# add new column for how long the flare lasted, in minutes
df2['length_mins'] = df2.apply(lambda x: \
  (x['end_datetime']-x['start_datetime']).seconds // 60, axis=1)

# add new column for whether the flare is in the top 50 flares 
# (in the SpaceWeatherLive dataset)
df2['is_top_50'] = df2['top_50_assoc_i'].map(lambda x: pandas.notna(x))

df2
```

```python
sns.set_theme(style="dark")

# actual plotting -- show the duration of a flare plotted
# over when it occurred, and use size to show the peak flux range.
# as requested in the README, the top 50 points are also marked.
g = sns.relplot(x='start_datetime', y='length_mins', data=df2,
                    size='peak_flux_range', sizes=(10,100),
                    hue='is_top_50', alpha=0.75,
                    height=6)

# set axis labels / title
g.set_axis_labels('Start Date','Duration of Flare (minutes)')
g.fig.suptitle('Duration of Flares Over Time',y=1.025, x=0.45)

# update text in legend -- remove title from hue category, 
# make sections more descriptive, move legend to not overlap graph
legend_text_labels = ['', 'Not In Top 50', 'In Top 50', 'Peak Flux Range (W/m²)']
for i,text in enumerate(legend_text_labels):
    g._legend.texts[i].set_text(text)
g._legend.set_bbox_to_anchor((1.1,0.85))
```
