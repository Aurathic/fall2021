```python
import folium
import requests
import pandas

arrest_table = pandas.read_csv("https://cmsc320.github.io/files/BPD_Arrests.csv")

arrest_table = arrest_table[pandas.notnull(arrest_table["Location 1"])]

arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str
arrest_table["lat"] = arrest_table["lat"].str.replace("(", "").astype(float)
arrest_table["long"] = arrest_table["long"].str.replace(")", "").astype(float)
```

## Part 3: Combining Parts 1 and 2

First, there are too many arrest to plot all of them. In order to reduce the number, we could look only at a portion of them -- say, the most recent two weeks of arrests. I chose to do this because ideally, the most recent data will be most reflective of the current data. (I fine-tuned this value until the visualization stopped making my computer lag.)

```python
import datetime

arrest_table["arrestDateObj"] = arrest_table["arrestDate"].apply(lambda row: datetime.datetime.strptime(row,"%m/%d/%Y"))
#arrest_table
```

```python
limitDate = datetime.date(2012, 12, 17)
arrest_table_limited = arrest_table[arrest_table['arrestDateObj'].dt.date > limitDate]
arrest_table_limited
```

953 entries is a much more manageable size than 63892 entries.

Note: Looking at these entries is useful if we want to analyze crime that occurs in this time frame, but we can't generalize what we see from this data to the whole two-year span. If we wanted to do that, we should instead take a random sample of the data (which should be distributed roughly equally across the time frame). 

In addition, I want to specifically indicate the type of crime with the icon, but there are too many types of incidents to give each their own indicator. In order to make the data comprehensible, we want to group together similar incidents under the same marker. 

First, it makes sense to give the most common incidents their own marker, so let's see which incidents are most commonly reported:

```python
arrest_groups_limited = arrest_table_limited.groupby('incidentOffense')
arrest_groups_limited.count().sort_values('arrest',ascending=False)['arrest']
```

Fortunately, it seems that every offense has a crime code if some sort assigned. 

An easy way to categorize all the items would be to give, say, the first 10 entries unique icons, then lumping the other types of crimes into a category 'Other (Categorized)'. To make these distinguishable, we can put each icon type on its own layer. To make it more useful, we can also add an on-hover tooltip that describes the offense in more detail.

(I initially had looked through the [crime codes](https://data.baltimorecity.gov/documents/crime-codes/about) and was going to group them together based on which crimes were most similar, but a majority of the offenses here only take up a small proportion of the total crimes, so it wouldn't be useful to try and group them together in this manner -- i.e. it doesn't help improve clarity.)  

```python
from pandas import isna
map_osm = folium.Map(location=[39.29, -76.61], zoom_start=11)


icons = [
    {'color': 'red', 'icon': 'minus-circle'},
    {'color': 'orange', 'icon': 'question'},
    {'color': 'beige', 'icon': 'hand-rock-o'},
    {'color': 'lightgreen', 'icon': 'search'},
    {'color': 'darkgreen', 'icon': 'ellipsis-h'},
    {'color': 'lightblue', 'icon': 'truck'},
    {'color': 'blue', 'icon': 'blind'},
    {'color': 'darkblue', 'icon': 'minus-circle'},
    {'color': 'purple', 'icon': 'cut'},
    {'color': 'pink', 'icon': 'ban'}
]

icon_other = {'color': 'gray', 'icon': 'ellipsis-v'}

d = dict(tuple(arrest_groups_limited))
offense_list = arrest_groups_limited.count().sort_values('arrest',ascending=False).index

# Note: if there is no given detailed description, I replace it with 'No description'
for offense,icon in zip(offense_list[:len(icons)],icons):
    fg = folium.FeatureGroup(offense)
    d[offense].apply(lambda row: folium.Marker(location=[row['lat'], row['long']], \
                                               tooltip=("No description" if isna(row['chargeDescription']) else row['chargeDescription']), \
                                     icon=folium.Icon(**icon, prefix='fa')).add_to(fg), axis=1)
    fg.add_to(map_osm)
    
# add all the other offenses under 'Other'
fg_other = folium.FeatureGroup('Other (Categorized)')
for offense in offense_list[len(icons):]:
    d[offense].apply(lambda row: folium.Marker(location=[row['lat'], row['long']], \
                                               tooltip=offense + " :: " + \
                                               ("No description" if isna(row['chargeDescription']) else row['chargeDescription']), \
                                     icon=folium.Icon(**icon_other, prefix='fa')).add_to(fg_other), axis=1)
    
fg_other.add_to(map_osm)

folium.LayerControl().add_to(map_osm)


map_osm
```

### Short Prose Summary
The data shown are all the crimes in the given database that occured in the final two weeks of 2012 (specifically between December 17th and December 31st, inclusive), giving insight into the distribution in that sliver of time. The data is split up into several layers based on the type of offense, as assigned by BPD's crime codes. The ten most common offenses are each given their own layer, including a distinct color pin and icon picture; the rest of them are lumped into a final miscellaneoyus category. (The icons are useful in case the viewer is color blind, as well as helping people get an idea what the offense category means without reading the layer controls.) 

There's a lot of stuff shown on the map, but you can focus on specific crimes (removing the clutter) by changing the layers that are visible in the layer control panel at the top right. In addition, if you hover over each of the specific offenses, you will see extra information about the offense, as well as their provided category if they are labelled as 'Other (Uncategorized)'.
