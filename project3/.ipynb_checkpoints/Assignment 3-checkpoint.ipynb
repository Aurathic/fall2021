## Part 1: Regression analysis of Gapminder data

```python
import seaborn as sns, numpy as np, matplotlib.pyplot as plt, sklearn as sk
```

```python
import pandas as pd
data = pd.read_csv("mirror/08_gap-every-five-years.tsv", sep='\t')
data.head()
```

**Exercise 1**: *Make a scatter plot of life expectancy across time.*

```python
ax = sns.scatterplot(y="lifeExp",x="year",data=data)
ax.set_xlabel("Year")
ax.set_ylabel("Life Expectancy (years)")
ax.set_title("Life Expectancy versus Year")
```

**Question 1**: *Is there a general trend (e.g., increasing or decreasing) for life expectancy across time? Is this trend linear? (answering this qualitatively from the plot, you will do a statistical analysis of this question shortly)*

There is a general positive (increasing) trend in life expectancy over time. It appears to be roughly linear.

```python
# Make a violin plot

fig, ax = plt.subplots()
sns.violinplot(x="year",y="lifeExp",data=data)
ax.set_xlabel("Year")
ax.set_ylabel("Life Expectancy (years)")
ax.set_title("Life Expectancy per Year")
```

**Question 2**: <em>How would you describe the distribution of life expectancy across countries for individual years? Is it skewed, or not? Unimodal or not? Symmetric around its [sic] center?</em>

For the years between 1952 and 1972 or so, the data appears skewed right (the tail points toward higher values). For the most recent years (since 1987 or so), the data appears skewed left. In between those years, the distribution is roughly symmetrical. \\
The early years in this sample seem bimodal, with the lower life expectancy peak being larger. Over time, however, the distribution became unimodal.

**Question 3**: <em>Suppose I fit a linear regression model of life expectancy vs. year (treating it as a continuous variable), and test for a relationship between year and life expectancy, will you reject the null hypothesis of no relationship? (do this without fitting the model yet. I am testing your intuition.)</em>

Since there is a visible positive correlation, you will likely reject the null hypothesis, though the relationship appears relatively weak so the p-value may not be very small.  

**Question 4**: <em>What would a violin plot of residuals from the linear model in Question 3 vs. year look like? (Again, dont do the analysis yet, answer this intuitively)</em>

Ideally, assuming that the data does follow a perfectly linear relationship, the violin plots would be symmmetrical in each year, all with the same variation. Furthermore, we would expect the medians of each year to fit onto a perfect line. However, since there is some skew shown in the violin plot, this will not be exactly the case. The earlier year distributions will be skewed upward, and the later year distributions will be skewed downward.

**Question 5**: <em>According to the assumptions of the linear regression model, what <strong>should</strong> that violin plot look like?  That is, consider the assumptions the linear regression model you used assumes (e.g., about noise, about input distributions, etc); do you think everything is okay?</em>

The assumptions of the linear regression model would expect that the residuals for each year should have symmetrical, unimodal distributions, all with the same standard deviation and all centered around 0. While our model does not fit that *exactly*, it seems fairly close, enough that I am satisfied with using a linear regression.

**Exercise 2**: <em>Fit a linear regression model using, e.g., the `LinearRegression` function from Scikit-Learn or the closed-form solution, for life expectancy vs. year (as a continuous variable).  There is no need to plot anything here, but please print the fitted model out in a readable format.</em>

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
# Turn columns into form that can be input into lin_reg
x = data["year"].values.reshape(-1,1)
y = data["lifeExp"].values.reshape(-1,1)
lin_reg.fit(x,y)
print(f"y ~ {lin_reg.intercept_} + {lin_reg.coef_[0]} * x")
```

**Question 6**: <em>On average, by how much does life expectancy increase every year around the world?</em>

For each single year increase, the model predicts that life expectancy will increase by approximately 0.33 years.

**Question 7**: <em>Do you reject the null hypothesis of no relationship between year and life expectancy? Why?</em>

We can see that the coefficient of the linear regression model is positive, but that alone is not enough information to conclude that we can reject the null hypothesis. We would need to get the variance or p-value, which we can calculate using statsmodels:

```python
import statsmodels.formula.api as smf
lin_reg = smf.ols("lifeExp ~ year", data=data).fit()
print(lin_reg.summary())
```

We can see the p-value for the slope is less than 0.001, which is certainly significant, so we reject the null hypothesis and say that there is a relationship.

**Exercise 3**: <em>Make a violin plot of residuals vs. year for the linear model from Exercise 2.</em>

```python
data["lifeExpResid"] = data["lifeExp"] - lin_reg.predict([[x] for x in data["year"]]).flatten()
data["lifeExpResid"]

# Violin plot 
fig, ax = plt.subplots()
sns.violinplot(x="year",y="lifeExpResid",data=data)
ax.set_xlabel("Year")
ax.set_ylabel("Life Expectancy Residual (years)")
ax.set_title("Life Expectancy Residuals per Year")
```

**Question 8**: <em>Does the plot of Exercise 3 match your expectations (as you answered Question 4)?</em>

The medians of each year range between roughly -5 and 5 from 1952 to 2007. There is a general trend that the earlier years have their life expectancy overestimated and the later years are underestimated. The distributions are still non-symmetric for the most part, which is also not expected given the assumption that the data is completely linear. In addition, as above, there are some bimodal distributions for some years, not unimodal. Therefore, some of my expectations were violated.

**Exercise 4**: <em>Make a boxplot (or violin plot) of model residuals vs. continent.</em>

```python
# Violin plot 
fig, ax = plt.subplots()
sns.violinplot(x="continent",y="lifeExpResid",data=data)
ax.set_xlabel("Continent")
ax.set_ylabel("Life Expectancy Residual (years)")
ax.set_title("Life Expectancy Residuals by Continent")
```

**Question 9**: <em>Is there a dependence between model residual and continent? If so, what would that suggest when performing a regression analysis of life expectancy across time?</em>

There is a clear dependence between model residual and continent. For example, the model consistent underestimates the life expectancy of those in Europe and Oceania, while overestimating that in Africa. This indicates that we should need to consider continent as a variable in our regression (or, alternatively, just do separate regressions on each continent's data separately).

**Exercise 5**: <em>As in the Moneyball project, make a scatter plot of life expectancy vs. year, grouped by continent, and add a regression line.  The result here can be given as either one scatter plot per continent, each with its own regression line, or a single plot with each continent's points plotted in a different color, and one regression line per continent's points.  The former is probably easier to code up.</em>

```python
for continent_data in data.groupby("continent"):
    fig, ax = plt.subplots()
    sns.regplot(x="year",y="lifeExp",data=continent_data[1])
    ax.set_xlabel("Year")
    ax.set_ylabel("Life Expectancy (years)")
    ax.set_title(f"Life Expectancy versus Year In {continent_data[0]}")
```

**Question 10**: <em>Based on this plot, should your regression model include an interaction term for continent <strong>and</strong> year? Why?</em>

Yes, including an interaction term for continent as well as year is useful because there are clear differences in the distributions between the continents, and we can see based on the previous exercise answer that there is a relationship between year and life expectancy. If a variable explains some of the life expectancy's variation, then it should be included, and both continent and year explain some variation.  

**Exercise 6**: <em>Fit a linear regression model for life expectancy including a term for an interaction between continent and year.  Print out the model in a readable format, e.g., print the coefficients of the model (no need to plot).  Hint: adding interaction terms is a form of feature engineering, like we discussed in class (think about, e.g., using (a subset of) polynomial features here).</em>

```python
import statsmodels.formula.api as smf
reg = smf.ols("lifeExp ~ year + year:continent", data=data).fit()
print(reg.summary())
```

**Question 11**: <em>Are all parameters in the model significantly (in the p-value sense) different from zero? If not, which are not significantly different from zero? Other libraries (`statsmodels` or `patsy` may help you solve this problem)</em>

In my the model above, all of the regressed variables have p-values which are less than 0.05, so they are all differ from 0 in a statistically significant manner.

**Question 12**: <em>On average, by how much does life expectancy increase each year for each continent? (Provide code to answer this question by extracting relevant estimates from model fit)</em>

```python
year_coef = reg.params["year"]
print("Life expectancy increase for:")
continents = ["Africa", "Americas", "Asia", "Europe", "Oceania"]
added_coefs = np.insert(reg.params.values[2:],0,0)
for (cont, coef) in zip(continents, added_coefs):
    print(f"{cont}: {year_coef+coef}")
```

**Exercise 7**: <em>Make a residuals vs. year violin plot for the interaction model. Comment on how well it matches assumptions of the linear regression model.</em>

```python
data["lifeExpResidInteraction"] = data["lifeExp"] - reg.predict(data)
fig, ax = plt.subplots()
sns.violinplot(x="year",y="lifeExpResidInteraction",data=data)
ax.set_xlabel("Year")
ax.set_ylabel("Life Expectancy Residual (years)")
ax.set_title("Life Expectancy Residuals (Including Continent Interactions)")
```

The means and medians for all of these years are even closer to 0 than for the regression that only considered year, and the variations are consistently smaller as well. Furthermore, unlike for the previous plot of residuals, all of the residuals are very close to symmetric, and the deviations for all of the residuals are very close to identical (though they spread out a tiny bit in later years). Overall, however, it seems like a very good match to the 'ideal' output from a distribution perfectly matching a linear regression model.

I chose the wisconsin breast cancer dataset, and used Linear Discriminant Analysis and decision trees to classify the data, then use 10-fold cross-validation to determine accuracy. The goal of this model is to classify a breast mass as either benign ('target' equals 1) or malignant ('target' equals 0) based on radius, texture, and other visible attributes.

```python
from sklearn.datasets import load_breast_cancer
data2 = load_breast_cancer(as_frame=True)
#print(data2.DESCR)

# I considered removing some of the variables which might be too similar to one another, but decided against it
x = data2["data"]
y = data2["frame"]["target"]
#print(len(y[y==1]))

data2["frame"]
# All of the data is already numeric, so no transformations are necessary
```

```python
# Linear Discriminant Analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import cross_validate

lda = LinearDiscriminantAnalysis()
validate_lda = cross_validate(lda, x, y, cv=10, scoring="accuracy", return_estimator=True)
#clfs_lda = validate_lda["estimator"]
scores_lda = validate_lda["test_score"]

print("accuracy mean: %0.4f; standard error: %0.4f" % (scores_lda.mean(), scores_lda.std()/np.sqrt(10)))
```

```python
# Decision Trees
from sklearn.tree import DecisionTreeClassifier

# I left the decision tree parameters the same as default (pick 'best' branch, use gini for measuring splits, et cetera),
# except I set the random state to be fixed in order to get reproducible results (removing randomness)
dt = DecisionTreeClassifier(random_state=0)
validate_dt = cross_validate(dt, x, y, cv=10, scoring="accuracy", return_estimator=True)
#clfs_dt = validate_dt["estimator"] 
scores_dt = validate_dt["test_score"]

print("accuracy mean: %0.4f; standard error: %0.4f" % (scores_dt.mean(), scores_dt.std()/np.sqrt(10)))
```

We can see that the linear discriminant analysis has both a higher accuracy by roughly 4 percent and a lower standard error among cross validated samples (by around 0.0025). 

```python
# Calculating paired t test between cross validation classifiers for each set of 2 classifiers
from scipy.stats import ttest_rel
ttest_rel(scores_lda, scores_dt)
```

The difference between the models is statistically significant (p<0.05), so we can say that there is a meaningful difference in the performance of the LDA and Decision Tree classifiers.
